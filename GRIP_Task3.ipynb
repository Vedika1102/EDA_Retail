```python
# lets see the counting 
plt.figure(figsize=(15,15))
plt.subplot(2,2,1)
sns.countplot(df["Ship Mode"],data=df)
plt.title("ship mode count")
plt.subplot(2,2,2)
sns.countplot(df["Segment"],data=df)
plt.title("Segment count")
plt.subplot(2,2,3)
sns.countplot(df["Region"],data=df)
plt.title("Region count")
plt.subplot(2,2,4)
sns.countplot(df["Category"],data=df)
plt.title("Category count")
```
